# Примеры предварительной обработки текста

```python
text_for_example = 'Зайку бросила хозяйка - Под дождем остался зайка. Со скамейки слезть не мог, Весь до ниточки промок.'
```

```python
import pandas as pd
```

### Примеры токенизации

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import sent_tokenize

sent_tokenize(text_for_example, language="russian")
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

tokens = word_tokenize(text_for_example, language="russian")
tokens
```

### Примеры удаления стоп-слов

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = stopwords.words("russian")

stop_words
```

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = stopwords.words("russian")

filtered_tokens = []

for token in tokens:
    if token not in stop_words:
        filtered_tokens.append(token)

filtered_tokens
```

### Пример стемминга

```python
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer(language="russian")

stemming = []

for token in filtered_tokens:
  stemming.append(snowball.stem(token))

diff_stemming = pd.DataFrame({
    'original': filtered_tokens,
    'w/stem': stemming,
}, columns=['original', 'w/stem'])

diff_stemming
```

### Пример лемматизации

```python
!pip install pymorphy2
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()

lemmatized = []

for token in filtered_tokens:
  lemmatized.append(morph.parse(token)[0].normal_form)

lemmatized

diff_lemm = pd.DataFrame({
    'original': filtered_tokens,
    'w/lemm': lemmatized,
}, columns=['original', 'w/lemm'])

diff_lemm
```

```python
diffs = pd.DataFrame({
    'original': filtered_tokens,
    'w/stem': stemming,
    'w/lemm': lemmatized,
}, columns=['original', 'w/stem', 'w/lemm'])

diffs
```

### Пример векторизации

```python
!pip install transformers
```

```python
from transformers import BertTokenizer

# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('blanchefort/rubert-base-cased-sentiment', 
                                          do_lower_case=False)
```

```python
encoded_sent = tokenizer.encode_plus(
            text=lemmatized,  # Preprocess sentence
            add_special_tokens=True,        # Add `[CLS]` and `[SEP]`
            max_length=256,                  # Max length to truncate/pad
            pad_to_max_length=True,         # Pad sentence to max length
            return_tensors='pt',           # Return PyTorch tensor
            return_attention_mask=True      # Return attention mask
            )
input_id = encoded_sent.get('input_ids')
attention_mask = encoded_sent.get('attention_mask')
```
